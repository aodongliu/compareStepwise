```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
from matplotlib import ticker
import h5py
import matplotlib as mpl
mpl.rc('text', usetex=True)
mpl.rcParams['text.latex.preamble']=[r"\usepackage{amsmath}"]
```

```python
eigen_stw_e = np.loadtxt("/Users/aodongliu/LiGroup/for_mathew/4h2o+/Data_new/eigen/eigen_stw_e.dat", dtype=None)
eigen_stw_pe = np.loadtxt("/Users/aodongliu/LiGroup/for_mathew/4h2o+/Data_new/eigen/eigen_stw_pe.dat", dtype=None)
eigen_stw_pp = np.loadtxt("/Users/aodongliu/LiGroup/for_mathew/4h2o+/Data_new/eigen/eigen_stw_pp.dat", dtype=None)
```

```python
eigen_smt_e = np.loadtxt("/Users/aodongliu/LiGroup/for_mathew/4h2o+/Data_new/eigen/eigen_smt_e.dat", dtype=None)
eigen_smt_pe = np.loadtxt("/Users/aodongliu/LiGroup/for_mathew/4h2o+/Data_new/eigen/eigen_smt_pe.dat", dtype=None)
eigen_smt_pp = np.loadtxt("/Users/aodongliu/LiGroup/for_mathew/4h2o+/Data_new/eigen/eigen_smt_pp.dat", dtype=None)
```

```python
eigen_stw_iter = np.arange(1,eigen_stw_e.size,1)
eigen_smt_iter = np.arange(1,eigen_smt_e.size,1)
```

```python
# Subtract the final energy for energy at each iteration
eigen_stw_e -= eigen_stw_e[-1]
eigen_smt_e -= eigen_smt_e[-1]
```

```python
plt.style.use('default')
fig = plt.figure(figsize=(12,8))
gs = fig.add_gridspec(3, hspace=0)
ax = gs.subplots(sharex=True)
#fig.suptitle('NEO-SCF with Different Algorithms, hatz/pb4d, $4H_2O+$ eigen Isomer')
ax[0].plot(eigen_smt_iter,eigen_smt_e[1:],'o-',c='darkorange',ms=2,label="Simultaneous-DIIS-Gau")
ax[0].plot(eigen_stw_iter,eigen_stw_e[1:],'o-',c='#9e66ab',ms=2,label="Stepwise-DIIS-Gau")
ax[0].axhline(y=10**(-10),xmin=0,xmax=eigen_stw_iter.size+1,c="black",linewidth=2,zorder=0,label='$E$ Threshold')
ax[0].set_yscale("log")
ax[0].set_ylabel("$ \Delta E}$",fontsize=25)
#ax[0].set_yticks([10**(-10),0])
#ax[0].legend(loc='upper right')
ax[0].spines['right'].set_visible(False)
ax[0].spines['top'].set_visible(False)

ax[1].plot(eigen_smt_iter,eigen_smt_pe[1:],'o-',c='darkorange',ms=2)
ax[1].plot(eigen_stw_iter,eigen_stw_pe[1:],'o-',c='#9e66ab',ms=2)
ax[1].axhline(y=10**(-8),xmin=0,xmax=eigen_stw_iter.size+1,c="red",linewidth=2,zorder=0,label='$RMS \Delta \mathbf{ \mathrm{P} } ^p$ Threshold')
ax[1].set_yscale("log")
ax[1].set_ylabel("$RMS \Delta \mathbf{P} ^e$",fontsize=25)
#ax[1].legend(loc='upper right')
ax[1].spines['right'].set_visible(False)

ax[2].plot(eigen_smt_iter,eigen_smt_pp[1:],'o-',c='darkorange',ms=2)
ax[2].plot(eigen_stw_iter,eigen_stw_pp[1:],'o-',c='#9e66ab',ms=2)
ax[2].axhline(y=10**(-8),xmin=0,xmax=eigen_stw_iter.size+1,c="red",linewidth=2,zorder=0)
ax[2].set_yscale("log")
ax[2].set_ylabel("$RMS \Delta \mathbf{P} ^p$",fontsize=25)
#ax[2].legend(loc='upper right')
ax[2].set_xlabel("Number of SCF Iterations", fontsize=25)
ax[2].set_xticks(np.arange(0, eigen_stw_iter.size+1, step=40))
ax[2].set_xticklabels(ax[2].get_xticks(), rotation=0, size=20)
ax[2].set_xlim(xmin=0)
ax[2].spines['right'].set_visible(False)

ax[0].tick_params(axis='y', labelsize=16)
ax[1].tick_params(axis='y', labelsize=16)
ax[2].tick_params(axis='y', labelsize=16)
# # Hide x labels and tick labels for all but bottom plot.
# for axes in ax:
#    axes.label_outer()

fig.set_facecolor('w')
lines_labels = [axes.get_legend_handles_labels() for axes in ax]
lines, labels = [sum(lol, []) for lol in zip(*lines_labels)]
fig.legend(lines, labels,fontsize=12,loc='upper right')

plt.tight_layout()
plt.savefig('/Users/aodongliu/LiGroup/for_mathew/plots/eigen_smt_vs_stw_fixed.pdf', dpi=3600)
# plt.show()



#bbox  = legend.get_window_extent().transformed(fig2.dpi_scale_trans.inverted())
#fig2.savefig(filename, dpi="figure", bbox_inches=bbox, facecolor="w")
        #legend.remove() # removes legend from showing on plot
```
