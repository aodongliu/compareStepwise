```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
from matplotlib import ticker
import h5py
import matplotlib as mpl
mpl.rc('text', usetex=True)
mpl.rcParams['text.latex.preamble']=[r"\usepackage{amsmath}"]
```

```python
zundel_stw_e = np.loadtxt("/Users/aodongliu/LiGroup/gaussian/DIIS_benchmark/4H2O+/zundel/zundel_stw_e.dat", dtype=None)
zundel_stw_pe_raw = np.loadtxt("/Users/aodongliu/LiGroup/gaussian/DIIS_benchmark/4H2O+/zundel/zundel_stw_pe.dat", dtype=None)
zundel_stw_pp_raw = np.loadtxt("/Users/aodongliu/LiGroup/gaussian/DIIS_benchmark/4H2O+/zundel/zundel_stw_pp.dat", dtype=None)

zundel_smtd_e = np.loadtxt("/Users/aodongliu/LiGroup/gaussian/DIIS_benchmark/4H2O+/zundel/zundel_smtd_e.dat", dtype=None)
zundel_smtd_pe = np.loadtxt("/Users/aodongliu/LiGroup/gaussian/DIIS_benchmark/4H2O+/zundel/zundel_smtd_pe.dat", dtype=None)
zundel_smtd_pp = np.loadtxt("/Users/aodongliu/LiGroup/gaussian/DIIS_benchmark/4H2O+/zundel/zundel_smtd_pp.dat", dtype=None)

zundelv_stw_e = np.loadtxt("/Users/aodongliu/LiGroup/gaussian/DIIS_benchmark/4H2O+/zundel/zundelv_stw_e.dat", dtype=None)
zundelv_stw_pe = np.loadtxt("/Users/aodongliu/LiGroup/gaussian/DIIS_benchmark/4H2O+/zundel/zundelv_stw_pe.dat", dtype=None)
zundelv_stw_pp = np.loadtxt("/Users/aodongliu/LiGroup/gaussian/DIIS_benchmark/4H2O+/zundel/zundelv_stw_pp.dat", dtype=None)

zundelv_smtd_e = np.loadtxt("/Users/aodongliu/LiGroup/gaussian/DIIS_benchmark/4H2O+/zundel/zundelv_smtd_e.dat", dtype=None)
zundelv_smtd_pe = np.loadtxt("/Users/aodongliu/LiGroup/gaussian/DIIS_benchmark/4H2O+/zundel/zundelv_smtd_pe.dat", dtype=None)
zundelv_smtd_pp = np.loadtxt("/Users/aodongliu/LiGroup/gaussian/DIIS_benchmark/4H2O+/zundel/zundelv_smtd_pp.dat", dtype=None)

qc_zundel_stw_e = np.loadtxt("/Users/aodongliu/LiGroup/for_mathew/QC_zundel/QC_zundel_stw_e.dat", dtype=None)
qc_zundel_stw_pe = np.loadtxt("/Users/aodongliu/LiGroup/for_mathew/QC_zundel/QC_zundel_stw_pe.dat", dtype=None)
qc_zundel_stw_pp = np.loadtxt("/Users/aodongliu/LiGroup/for_mathew/QC_zundel/QC_zundel_stw_pp.dat", dtype=None)
```

```python
lenpe = zundel_stw_pe_raw.size
zundel_stw_pe = []
last_val = zundel_stw_pe_raw[0]
zundel_stw_pe.append(last_val)
index_to_keep = []
for i in range(1,lenpe):
    cur_val = zundel_stw_pe_raw[i]
    if ( not(last_val  >= 10**(-8) and cur_val  <= 10**(-8)) ):
        zundel_stw_pe.append(cur_val)
        index_to_keep.append(i)
    last_val = cur_val
```

```python
lenpp = zundel_stw_pp_raw.size
zundel_stw_pp = []
last_val = zundel_stw_pp_raw[0]
zundel_stw_pp.append(last_val)
for i in range(1,lenpp):
    cur_val = zundel_stw_pp_raw[i]
    if ( i in index_to_keep):
        zundel_stw_pp.append(cur_val)
```

```python
zundel_stw_iter = np.arange(1,zundel_stw_e.size,1)
zundel_smtd_iter = np.arange(1,zundel_smtd_e.size,1)
zundelv_smtd_iter = np.arange(1,zundelv_smtd_e.size,1)
zundelv_stw_iter = np.arange(1,zundelv_stw_e.size,1)
qc_zundel_stw_iter = np.arange(1,qc_zundel_stw_e.size,1)
```

```python
# Subtract the final energy for energy at each iteration
zundel_stw_e -= zundel_stw_e[-1]
zundel_smtd_e -= zundel_smtd_e[-1]
zundelv_stw_e -= zundelv_stw_e[-1]
zundelv_smtd_e -= zundelv_smtd_e[-1]
qc_zundel_stw_e -= qc_zundel_stw_e[-1]
```

```python
plt.style.use('default')
fig = plt.figure(figsize=(12,8))
gs = fig.add_gridspec(3, hspace=0)
ax = gs.subplots(sharex=True)
#fig.suptitle('NEO-SCF with Different Algorithms, hatz/pb4d, $4H_2O+$ zundel Isomer')
ax[0].plot(zundel_smtd_iter,zundel_smtd_e[1:],'o-',c='darkorange',ms=2,label="Simultaneous-DIIS-Gau")
ax[0].plot(zundel_stw_iter,zundel_stw_e[1:],'o-',c='#9e66ab',ms=2,label="Stepwise-DIIS-Gau")
ax[0].plot(zundelv_smtd_iter,zundelv_smtd_e[1:],'o-',c='g',ms=2,label="Simultaneous-DIIS-Gau-V")
ax[0].plot(zundelv_stw_iter,zundelv_stw_e[1:],'o-',c='y',ms=2,label="Stepwise-DIIS-Gau-V")
ax[0].plot(qc_zundel_stw_iter,qc_zundel_stw_e[1:],'o-',c='#599ad3',ms=2,label="Stepwise-DIIS-QC")
ax[0].axhline(y=10**(-10),xmin=0,xmax=zundel_stw_iter.size+1,c="black",linewidth=2,zorder=0,label='$E$ Threshold')
ax[0].set_yscale("log")
ax[0].set_ylabel("$ \Delta E}$",fontsize=25)
#ax[0].set_yticks([10**(-10),0])
#ax[0].legend(loc='upper right')
ax[0].spines['right'].set_visible(False)
ax[0].spines['top'].set_visible(False)

ax[1].plot(zundel_smtd_iter,zundel_smtd_pe[1:],'o-',c='darkorange',ms=2)
ax[1].plot(zundel_stw_iter,zundel_stw_pe[1:],'o-',c='#9e66ab',ms=2)
ax[1].plot(zundelv_smtd_iter,zundelv_smtd_pe[1:],'o-',c='g',ms=2)
ax[1].plot(zundelv_stw_iter,zundelv_stw_pe[1:],'o-',c='y',ms=2)
ax[1].plot(qc_zundel_stw_iter,qc_zundel_stw_pe[1:],'o-',c='#599ad3',ms=2)
ax[1].axhline(y=10**(-8),xmin=0,xmax=zundel_stw_iter.size+1,c="red",linewidth=2,zorder=0,label='$RMS \Delta \mathbf{ \mathrm{P} } ^p$ Threshold')
ax[1].set_yscale("log")
ax[1].set_ylabel("$RMS \Delta \mathbf{P} ^e$",fontsize=25)
#ax[1].legend(loc='upper right')
ax[1].spines['right'].set_visible(False)

ax[2].plot(zundel_smtd_iter,zundel_smtd_pp[1:],'o-',c='darkorange',ms=2)
ax[2].plot(zundel_stw_iter,zundel_stw_pp[1:],'o-',c='#9e66ab',ms=2)
ax[2].plot(zundelv_smtd_iter,zundelv_smtd_pp[1:],'o-',c='g',ms=2)
ax[2].plot(zundelv_stw_iter,zundelv_stw_pp[1:],'o-',c='y',ms=2)
ax[2].plot(qc_zundel_stw_iter,qc_zundel_stw_pp[1:],'o-',c='#599ad3',ms=2)
ax[2].axhline(y=10**(-8),xmin=0,xmax=zundel_stw_iter.size+1,c="red",linewidth=2,zorder=0)
ax[2].set_yscale("log")
ax[2].set_ylabel("$RMS \Delta \mathbf{P} ^p$",fontsize=25)
#ax[2].legend(loc='upper right')
ax[2].set_xlabel("Number of SCF Iterations", fontsize=25)
ax[2].set_xticks(np.arange(0, zundel_stw_iter.size+1, step=400))
ax[2].set_xticklabels(ax[2].get_xticks(), rotation=0, size=20)
ax[2].set_xlim(xmin=0)
ax[2].spines['right'].set_visible(False)

ax[0].tick_params(axis='y', labelsize=16)
ax[1].tick_params(axis='y', labelsize=16)
ax[2].tick_params(axis='y', labelsize=16)
# # Hide x labels and tick labels for all but bottom plot.
# for axes in ax:
#    axes.label_outer()

fig.set_facecolor('w')
lines_labels = [axes.get_legend_handles_labels() for axes in ax]
lines, labels = [sum(lol, []) for lol in zip(*lines_labels)]
fig.legend(lines, labels,fontsize=12,loc='upper right')

plt.tight_layout()
plt.savefig('/Users/aodongliu/LiGroup/for_mathew/plots/zundel_qc_vs_gau.pdf', dpi=3600)
# plt.show()



#bbox  = legend.get_window_extent().transformed(fig2.dpi_scale_trans.inverted())
#fig2.savefig(filename, dpi="figure", bbox_inches=bbox, facecolor="w")
        #legend.remove() # removes legend from showing on plot
```
